Simulating medical X-ray classification using CIFAR-10 (as required by module brief)

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

print("=== TOPIC 1 Part A: EDA – Human-Guided Only ===")

# Load dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

print(f"Train: {x_train.shape} | Test: {x_test.shape}")
print(f"Balanced classes: {len(np.unique(y_train))}")
```

```python
# Sample images (save for report)
plt.figure(figsize=(12,12))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i])
    plt.title(class_names[y_train[i][0]])
    plt.axis('off')
plt.suptitle("Topic 1: Sample Images – Visual Ambiguity (mirrors medical X-ray overlap)")
plt.savefig('reports/figures/sample_images.png')
plt.show()
```

```python
# Class distribution
plt.figure(figsize=(10,5))
sns.countplot(x=y_train.flatten())
plt.title("Class Distribution – Perfectly Balanced")
plt.savefig('reports/figures/class_distribution.png')
plt.show()
```

```python
# Pixel intensity
plt.hist(x_train.flatten(), bins=50)
plt.title("Pixel Intensity Distribution")
plt.savefig('reports/figures/pixel_histogram.png')
plt.show()

print("✅ EDA complete! All figures saved to reports/figures/")
print("Key insight for report: High class overlap (cat/dog, bird/plane) perfectly simulates ambiguous medical X-ray cases.")
```
